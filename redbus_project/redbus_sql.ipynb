```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project/df_buses_1.csv")
df_buses_2=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project/df_buses_2.csv")
df_buses_3=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_3.csv")
df_buses_4=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_4.csv")
df_buses_5=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_5.csv")
df_buses_6=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\/df_buses_6.csv")
df_buses_7=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_7.csv")
df_buses_8=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_8.csv")
df_buses_9=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_9.csv")
df_buses_10=pd.read_csv(r"C:\Users\USER\Desktop\redbus_project\df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

Data cleaning

```python
# data about the data
Final_df.info()
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
# change dataframe to csv
path=r"C:\Users\USER\Desktop\redbus_project/Final_busdetails_df.csv"
Final_df.to_csv(path,index=False)
```

Create a mysql database

```python
#sql connection
conn=mysql.connector.connect(host="localhost", user="root", password="nandu17@",database="RED_BUS_DETAILS")
my_cursor = conn.cursor()
my_cursor.execute("CREATE DATABASE IF NOT EXISTS RED_BUS_DETAILS")
```

```python
# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS bus_details(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Bus_name VARCHAR(255) NOT NULL,
                  Bus_type VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL,
                  Ratings Float NULL,
                  Route_link VARCHAR(255) NULL,
                  Route_name VARCHAR(255) NULL
                  )''')
print("Table Created successfully")
```
